```python
import io #(imput e output, entrada e saída). Para que o python possa manejar arquivos de entrada e saída de diferentes modos
import pandas as pd #pandas para manejar arquivos excel ou de tabelas
import geopandas as gpd

# Ler os dados CENTROIDE
centro=gpd.read_file('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/Organizacao_Dados/Area_Plantada_Bunge/Centroide_CAR_MT.shp')
```

```python
centro.head()
```

```python
# Ler os dados CAR
car_mt=gpd.read_file('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/Organizacao_Dados/Area_Plantada_Bunge/MT.shp')
```

```python
car_mt.head()
```

```python
# Ler os dados GRID
grid=gpd.read_file('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/grid_monitoramento.shp')
```

```python
grid.head()
```

```python
# Projeção grid
grid.crs
```

```python
# Projeção centroides
centro.crs
```

```python
# Tudo está no mesmo sistema de coordenadas, então pode fazer o spatial join

# Realizar o spatial join (sjoin) baseado no operador within. Ou seja,
# vamos juntar os centróides aos grids nos quais eles estão dentro
centro_join=gpd.sjoin(centro,grid,how='inner',predicate='within')
#how='inner' vamos manter só os que tem em comum nos 2 / se o objetivo fosse ter todos os censitários seria right (pq ele tá depois)
# predicate='whithin' (operador espacial) relação dos pontos dos bares e os setores, o ponto dos bares estão dentro dos setores
```

```python
# Verificar os primeiros registros
centro_join.head()
```

```python
# Selecionar apenas as colunas interessantes
# essa é uma forma de selecionar multiplas colunas de um lugar
centro_join=centro_join[['codigo','nome']]
centro_join.head()
```

```python
# Juntar o atributo nome (GRID) no shape de CAR do MT de acordo com o código "código"

# Verificar se tem o mesmo tipo de classe de dado
# Checar o tipo de dados da primeira linha da coluna codigo
print(type(centro_join.codigo[0]))

# Checar o tipo de dados da primeira linha da coluna codigo
print(type(car_mt.codigo[0]))
```

```python
# Já que os dados tem a mesma classe podemos fazer o table join
# Realizar o table join com base na coluna codigo - merge
car_grid=car_mt.merge(centro_join, on='codigo', how='left') #on= coluna chave para o table join
                              #how='left' todos os registros da tabela da esquerda (car_mt, porque vem primeiro) serão mantidos
```

```python
# Checar os primeiros registros da tabela resultando do table join
car_grid.head()
```

```python
# Exportar o shape do CAR de MT com grid do monitoramento
car_grid.to_file('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/Organizacao_Dados/Area_Plantada_Bunge/CAR_MT_com_grid.shp')
```
