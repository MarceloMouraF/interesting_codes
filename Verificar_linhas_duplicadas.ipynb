```python
# Carregue o arquivo Excel em um DataFrame, pulando a primeira linha
# A segunda linha vai ser o cabeçalho
planilha1 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/20230906/CARs Area de Estudo.xlsx', header=1)

planilha2 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/20230906/CARs Raio 30km.xlsx', header=1)
```

```python
planilha1.head(3)
```

```python
planilha_geral = pd.concat([planilha1,planilha2])

planilha_geral['duplicado'] = planilha_geral.duplicated(subset='CAR', keep=False)
```

```python
planilha_de_duplicados = planilha_geral[planilha_geral['duplicado']==True]

planilha_de_duplicados = planilha_de_duplicados.drop_duplicates(subset=['CAR'],keep='first')
```

```python
#planilha_de_duplicados.to_excel('/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/VEGA/20230906/CARs_duplicados.xlsx',index=False)
```

```python
# Excluir os atributos duplicados da planilha do Buffer (CARs que são duplicados, tem nas 2 planilhas)

# Faça um merge nas duas planilhas com base nos atributos comuns (chave primária)
# Suponha que 'atributo_comum' seja o nome do atributo que é comum em ambas as planilhas
merged = pd.merge(planilha2, planilha1, on='CAR', how='left', indicator=True)

# Filtrar as linhas da Planilha 2 que não têm correspondência na Planilha 1
planilha2_sem_duplicatas = merged[merged['_merge'] == 'left_only'].drop(columns=['_merge'])
```

```python
# Verificar quantos foram excluidos da planilha 2
print(planilha2.shape)
print(planilha2_sem_duplicatas.shape)
```

```python
# Ver as colunas que a planilha2_sem_duplicatas tem
planilha2_sem_duplicatas.columns
```

```python
# Selecionar as colunas desejadas da planilha2_sem_duplicatas
#seleciona colunas específicas
planilha2_final = planilha2_sem_duplicatas[['CAR','Área CAR Calculada_x', '% do CAR dentro do Raio de 30km',
                                            'Area do Car Dentro do Raio de 30km',
       'Situação_x', 'Condição_x','UF_x', 'municÍpio_x', 'ID municÍpio_x', 'Módulos Fiscais_x', 'Tipo_x',
       'Área (ha)_x', 'CEP_x']]
#renomeia a coluna
planilha2_final = planilha2_final.rename(columns={'uf_x':'uf','Área CAR Calculada_x':'Área CAR Calculada',
       'Situação_x':'Situação', 'Condição_x':'Condição','UF_x':'UF', 'municÍpio_x':'municÍpio', 'ID municÍpio_x':'ID municÍpio',
                                      'Módulos Fiscais_x':'Módulos Fiscais', 'Tipo_x':'Tipo','Área (ha)_x':'Área (ha)', 'CEP_x':'CEP'})
```
