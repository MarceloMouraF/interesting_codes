```python
# Mudar a ordem das coordenadas
pmilho_1=Point(-80.159766,43.174322)
pmilho_2=Point(-80.180328,43.168862)
pmilho_3=Point(-80.418961,43.153595)
pmilho_4=Point(-80.753789,43.099877)
psoja_1=Point(-80.418961,43.153595)
pmilho_5=Point(-80.778074,43.082592) # perto da ingersoil
pmilho_e_soja=Point(-80.060539,43.196304) #-80.06035,43.19621
```

**Criar um GeoDataBase vazio para colocar os pontos**

```python
geodf=gpd.GeoDataFrame()

# Criar a coluna geometry vazia
geodf['geometry']=None
geodf
```

```python
print(coords_whats)
```

```python
geodf['geometry']=coords_whats
geodf
```

```python
# Verificar o sistema de coordenadas
print(geodf.crs)

# Como não há sistema de coordenadas temos que atribuir um. Ele será: (Lat Long WGS84)
geodf=geodf.set_crs('epsg:4326')
# Esse sistema de coordenadas é usado no google earth e também no Canadá
# Verificar o sistema de coordenadas atribuido
print(geodf.crs)
geodf.crs
```

```python
# Adicionar outra coluna no geodataframe, como a maioria é milho vou atribuir milho para todos
geodf['class']='milho'
```

```python
# Ver o shape todo, como ele só tem 7 itens...
geodf.head(7)
```

```python
# Mudar os pontos que acho que não é milho
geodf.loc[5,'class']='soja'
geodf.loc[6,'class']='milho de um lado e soja do outro'
# Verificar os atributos de novo
geodf.head(7)
```

```python
geodf.crs
```

```python
# Exportar o GDF para um shapefile
# Para isso, precisamos criar uma pasta de saída. Então é necessário importar a biblioteca os (biblioteca para caminhos no Python)
import os

# Definir a pasta de saída
saida=r'/content/drive/MyDrive/Colab Notebooks/PyGEO/Pratica/Canada'

# Criar o caminho para a pasta onde serão salvos os shapefiles
pasta_resultado=os.path.join(saida,'Pratica/Canada')

#Criar uma nova pasta chamada Exercicio_M2 (caso ela não exista) usando a função os.makedirs()
if not os.path.exists(pasta_resultado):
  os.makedirs(pasta_resultado)

# Realmente exportar o GDF para um shapefile
geodf.to_file(r'/content/drive/MyDrive/Colab Notebooks/PyGEO//Pratica/Canada/gdf_coords_estrada_Canada.shp')

# OU
#nome_saida='gdf_coords.shp'
#geodf_coords.to_file(os.path.join(pasta_resultado,nome_saida))
```

```python
# Visualizar o shape
import folium

# Calcular a extensão do shape para centralizar o mapa
center = [geodf.geometry.centroid.y.mean(), geodf.geometry.centroid.x.mean()]

# Criar um mapa do Google Maps com folium
m = folium.Map(location=center, zoom_start=12, tiles='Stamen Terrain')

# Plotar o shape no mapa
folium.GeoJson(geodf).add_to(m)

# Exibir o mapa
m
```
